```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('epl-allseasons-matchstats.csv')
data
```

```python
y=data['HomeGoals']
x1 =data['HomeGoalsHalfTime']
plt.scatter(x1,y)
plt.xlabel('HomeGoalsHalfTime', fontsize=20)
plt.ylabel('HomeGoals',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
```

```python
results = sm.OLS(y,x).fit()
results.summary()
```
